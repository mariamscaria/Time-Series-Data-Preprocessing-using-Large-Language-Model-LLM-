```python
# write your prompt using the agent
import os
import pandas as pd
import numpy as np

# Setup the API key
os.environ['OPENAI_API_KEY'] = "your-api-key"

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/nimikaringada/Downloads/Hospital_patients_datasets.csv')

# Convert 'ScheduledDay' and 'AppointmentDay' columns to datetime objects, focusing only on the date part
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.date
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.date

# Rename columns
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'SMS_received': 'SMSReceived',
    'No-show': 'NoShow'
}, inplace=True)

# Drop columns
df.drop(columns=['PatientId', 'AppointmentID', 'Neighbourhood'], inplace=True)

# Remove rows where 'Age' equals 0
df = df[df['Age'] != 0]

# Create 'Age_group' column by categorizing ages into bins of 20 years
df['Age_group'] = pd.cut(df['Age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])

# Drop the 'Age' column
df.drop(columns=['Age'], inplace=True)

# Convert 'NoShow' column to binary values: 1 for 'Yes' and 0 for other values
df['NoShow'] = np.where(df['NoShow'] == 'Yes', 1, 0)

# Export the cleaned DataFrame
df.to_csv('cleaned6.csv', index=False)
df
```
